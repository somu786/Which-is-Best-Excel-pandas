```python
# Find which/what is best by giving file path of that excel.
# This Algorithm give Importance to Consistence over Average. (It will show which data is consistent and which data have high average)

#Sample input & output:
"""  
    press 1 for csv file, press 2 for excel file: 2
    
    File Path: C:/Users/91805/Documents/joeVskane.xlsx

    No.of Column to compare: 2

    Column 1 Name: root

    Column 2 Name: kane

    Result: 

     ----------------------------------------------------------------
    | root is Better with Consistent! but kane have higher Average!! |
     ----------------------------------------------------------------
"""
#Code:    
import pandas as pd
import statistics

#csv = input("File Name: ")  #Input File Name without .xlsx, Example->(JoeVsKane)
#data = pd.read_excel("C:/Users/91805/Documents/"+csv+".xlsx")  #Here excel sheet was imported by fileName(csv). Note: Here user no need to enter full directory path, instead user can enter only filename. this step will automatically finds the path.

flag=1    
while(flag):    
    choice = int(input("press 1 for csv file, press 2 for excel file: "))
    if(choice==1):  
        csv = input("File Path: ")  # Input -> File Path of the csv file
        data = pd.read_excel(csv)
        flag=0
    elif(choice==2):    
        excel = input("File Path: ") # Input -> File Path of the excel sheet
        data = pd.read_excel(excel) # above commented step did not ask user to enter their full path of the file, it takes input of FileName(document path is already initialized) but it is only works on my(91805) system. so it is not work on others . so user need to enter the path whichh contains the file(Excel). refer above sample input 
        flag=0
    else:
        print("Invalid")
   
noOfColumn = int(input("No.of Column to compare: ")) # how many columns to compare
#print()
column = []  # column name entered by user was stored in this list..
for i in range(1,noOfColumn+1):
    e = input("Column "+str(i)+" Name: ") #Column name input
    column.append(e) # column name was added to list 
print()    
sdValue = data[column[0]].std()  #Set initial standard value, after it will change, if it was low in another column
avg = [] #average value was stored in this list
for i in column:
    avg.append(data[i].mean())     # Average of each column will added to list 
    if(sdValue >= data[i].std()):    #comparing initial sd value with current data sd value
        sdValue = data[i].std()      # if current sd value is lower than what we initially setted , then current lower value is set to sdValue
        sdResult = i      # here we assign the data name in sdResult, which have sdValue as low 
print("Result: ")     
print()
if(sdResult == column[avg.index(max(avg))]): # if a data with high average and consistence then the bif condition will execute. 
    print(" "+"-"*len(sdResult)+"-"*43)
    print("| "+sdResult,"is Better with Consistent! and Average!! |")
    print(" "+"-"*len(sdResult)+"-"*43)
else: # consistence but high average
    print(" "+"-"*(len(sdResult)+len(column[avg.index(max(avg))]))+"-"*56)
    print("| "+sdResult,"is Better with Consistent! but",column[avg.index(max(avg))],"have higher Average!! |")   
    print(" "+"-"*(len(sdResult)+len(column[avg.index(max(avg))]))+"-"*56)
print()
print("Measures of central tendency:")
print("For",column[0],":")
print("   Mean:",data[column[0]].mean()) 
print("   Median:",data[column[0]].median()) 
print("   Mode:",statistics.mode(data[column[0]]))
print()
print("For",column[1],":")
print("   Mean:",data[column[1]].mean()) 
print("   Median:",data[column[1]].median()) 
print("   Mode:",statistics.mode(data[column[1]]))
```
